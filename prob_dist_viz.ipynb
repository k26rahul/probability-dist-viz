# Probability distributions

```python
import numpy as np
import matplotlib.pyplot as plt
```

# $X \sim \text{Bernoulli}(p)$

Models a single Bernoulli trial: success occurs with probability $p$, failure with probability $1-p$.

**Support:** $\{0, 1\}$  
**Expected Value:** $E[X] = p$  
**Variance:** $\text{Var}(X) = p(1-p)$

```python
p = 0.5         # probability of success in a single Bernoulli trial
repeat = 1000   # number of times we repeat the experiment to get multiple samples

values = np.random.choice([0, 1], size=repeat, p=[1-p, p])

print(f"First 5 values: {values[:5]}")
print(f"Mean: {values.mean():.4f}, Expected Mean: {p}")
print(f"Variance: {values.var():.4f}, Expected Variance: {p * (1 - p)}")

counts, bins = np.histogram(values, bins=[-0.5, 0.5, 1.5], density=True)
_ = plt.bar([0, 1], counts, color=['skyblue', 'salmon'], width=0.6)
_ = plt.xlabel('Value')
_ = plt.ylabel('Probability mass')
```

# $X \sim \text{Binomial}(n, p)$

Models the number of successes in $n$ independent Bernoulli($p$) trials.

**Support:** $\{0, 1, 2, \dots, n\}$  
**Expected Value:** $E[X] = n \cdot p$  
**Variance:** $\text{Var}(X) = n \cdot p \cdot (1-p)$

```python
n = 10          # number of Bernoulli trials in each experiment
p = 0.4         # probability of success
repeat = 1000   # number of times we repeat the entire n-trial experiment

values = np.random.binomial(n, p, size=repeat)

print(f"First 5 values: {values[:5]}")
print(f"Mean: {values.mean():.2f}, Expected Mean: {n*p:.2f}")
print(f"Variance: {values.var():.2f}, Expected Variance: {n*p*(1-p):.2f}")

bins = np.arange(values.min(), values.max() + 1) - 0.5
_ = plt.hist(values, bins=bins, density=True, color='skyblue', edgecolor='black')
_ = plt.xlabel('Number of successes')
_ = plt.ylabel('Probability mass')
```

# $X \sim \text{Geom}(p)$

Models the number of failures before the first success in repeated independent Bernoulli($p$) trials.

**Support:** $\{0, 1, 2, \dots\}$  
**Expected Value:** $E[X] = \frac{1-p}{p}$  
**Variance:** $\text{Var}(X) = \frac{1-p}{p^2}$

```python
p = 0.5         # probability of success in each Bernoulli trial
repeat = 1000   # number of times we repeat the experiment

# numpy returns trial number of first success, subtract 1 to get number of failures
values = np.random.geometric(p, size=repeat) - 1

print(f"First 5 values: {values[:5]}")
print(f"Mean: {values.mean():.2f}, Expected Mean: {(1-p)/p:.2f}")
print(f"Variance: {values.var():.2f}, Expected Variance: {(1-p)/p**2:.2f}")

bins = np.arange(values.min(), values.max() + 1) - 0.5
_ = plt.hist(values, bins=bins, density=True, color='skyblue', edgecolor='black')
_ = plt.xlabel('Number of failures before first success')
_ = plt.ylabel('Probability mass')
```

# $X \sim \text{NegBin}(r, p)$

Models the number of failures before observing $r$ successes in repeated independent Bernoulli($p$) trials.

**Support:** $\{0, 1, 2, \dots\}$  
**Expected Value:** $E[X] = \frac{r (1-p)}{p}$  
**Variance:** $\text{Var}(X) = \frac{r (1-p)}{p^2}$

```python
r = 10         # number of successes to observe
p = 0.5        # probability of success in each Bernoulli trial
repeat = 1000  # number of times we repeat the experiment

# numpy already returns number of failures before r-th success
values = np.random.negative_binomial(r, p, size=repeat)

print(f"First 5 values: {values[:5]}")
print(f"Mean: {values.mean():.2f}, Expected Mean: {(r*(1-p)/p):.2f}")
print(f"Variance: {values.var():.2f}, Expected Variance: {(r*(1-p)/p**2):.2f}")

bins = np.arange(values.min(), values.max() + 1) - 0.5
_ = plt.hist(values, bins=bins, density=True, color='skyblue', edgecolor='black')
_ = plt.xlabel('Number of failures before r-th success')
_ = plt.ylabel('Probability mass')
```

# $X \sim \text{Uniform}(a, b)$

Models a random variable that is equally likely to take any value in the interval $[a, b]$.

**Support:** $[a, b]$  
**Probability Density Function (PDF):** $f_X(x) = \frac{1}{b-a}, \quad a \le x \le b$  
**Expected Value:** $E[X] = \frac{a + b}{2}$  
**Variance:** $\text{Var}(X) = \frac{(b - a)^2}{12}$

```python
a = 0           # lower bound of uniform interval
b = 10          # upper bound of uniform interval
repeat = 1000   # number of times we sample

values = np.random.uniform(a, b, size=repeat)

print(f"First 5 values: {values[:5]}")
print(f"Mean: {values.mean():.2f}, Expected Mean: {(a+b)/2:.2f}")
print(f"Variance: {values.var():.2f}, Expected Variance: {((b-a)**2)/12:.2f}")

_ = plt.hist(values, bins=30, density=True, color='skyblue', edgecolor='black')
_ = plt.xlabel('Value')
_ = plt.ylabel('Probability density')
```

# $X \sim N(\mu, \sigma^2)$

Models a random variable with a bell-shaped, symmetric distribution around mean $\mu$ and variance $\sigma^2$.

**Support:** $(-\infty, \infty)$  
**Probability Density Function (PDF):**  
$f_X(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\Big(-\frac{(x-\mu)^2}{2\sigma^2}\Big)$  
**Expected Value:** $E[X] = \mu$  
**Variance:** $\text{Var}(X) = \sigma^2$

```python
mu = 100        # mean of the normal distribution
sigma = 50      # standard deviation
repeat = 1000   # number of samples

values = np.random.normal(mu, sigma, size=repeat)

print(f"First 5 values: {values[:5]}")
print(f"Mean: {values.mean():.2f}, Expected Mean: {mu:.2f}")
print(f"Variance: {values.var():.2f}, Expected Variance: {sigma**2:.2f}")

_ = plt.hist(values, bins=30, density=True, color='skyblue', edgecolor='black')
_ = plt.xlabel('Value')
_ = plt.ylabel('Probability density')
```

# $X \sim \text{Exponential}(\lambda)$

Models the waiting time until the first event occurs in a Poisson process with rate $\lambda$.

**Support:** $[0, \infty)$  
**Probability Density Function (PDF):**  
$f_X(x) = \lambda e^{-\lambda x}, \quad x \ge 0$  
**Expected Value:** $E[X] = \frac{1}{\lambda}$  
**Variance:** $\text{Var}(X) = \frac{1}{\lambda^2}$

```python
 = 1          # rate parameter (events per unit time)
repeat = 1000  # number of samples

values = np.random.exponential(1/, size=repeat)

print(f"First 5 values: {values[:5]}")
print(f"Mean: {values.mean():.2f}, Expected Mean: {1/:.2f}")
print(f"Variance: {values.var():.2f}, Expected Variance: {(1/**2):.2f}")

_ = plt.hist(values, bins=30, density=True, color='skyblue', edgecolor='black')
_ = plt.xlabel('Value')
_ = plt.ylabel('Probability density')
```

# $X \sim \text{Beta}(\alpha, \beta)$

Models a random variable taking values in $[0, 1]$, often used for probabilities or proportions.  

**Support:** $[0, 1]$  
**Probability Density Function (PDF):**  
$f_X(x) = \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)} x^{\alpha-1} (1-x)^{\beta-1}, \quad 0 \le x \le 1$  
**Expected Value:** $E[X] = \frac{\alpha}{\alpha+\beta}$  
**Variance:** $\text{Var}(X) = \frac{\alpha \beta}{(\alpha+\beta)^2 (\alpha+\beta+1)}$

```python
alpha = 2      # first shape parameter
beta = 5       # second shape parameter
repeat = 1000  # number of samples

values = np.random.beta(alpha, beta, size=repeat)

print(f"First 5 values: {values[:5]}")
print(f"Mean: {values.mean():.2f}, Expected Mean: {alpha/(alpha+beta):.2f}")
print(f"Variance: {values.var():.2f}, Expected Variance: {(alpha*beta)/((alpha+beta)**2*(alpha+beta+1)):.2f}")

_ = plt.hist(values, bins=30, density=True, color='skyblue', edgecolor='black')
_ = plt.xlabel('Value')
_ = plt.ylabel('Probability density')
```
