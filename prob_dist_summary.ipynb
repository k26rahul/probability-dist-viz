## Discrete Distributions

| Distribution | Support | PMF | Mean | Variance |
|---------------|----------|-----------|-----------|-----------|
| Bernoulli $(p)$ | $\{0,1\}$ | $p^x(1-p)^{1-x}$ | $p$ | $p(1-p)$ |
| Binomial $(n,p)$ | $\{0,1,2,\dots,n\}$ | $\displaystyle \binom{n}{k} p^k (1-p)^{n-k}$ | $np$ | $np(1-p)$ |
| Geometric $(p)$ | $\{1,2,\dots\}$ or $\{0,1,2,\dots\}$ | $(1-p)^{k-1}p$ or $(1-p)^k p$ | $1/p$ or $(1-p)/p$ | $(1-p)/p^2$ |
| Negative Binomial $(r,p)$ | $\{r,r+1,\dots\}$ or $\{0,1,2,\dots\}$ | $\displaystyle \binom{k-1}{r-1}p^r(1-p)^{k-r}$ or $\displaystyle \binom{k+r-1}{r-1}p^r(1-p)^k$ | $r/p$ or $r(1-p)/p$ | $r(1-p)/p^2$ |
| Poisson $(\lambda)$ | $\{0,1,2,\dots\}$ | $\displaystyle \frac{e^{-\lambda}\lambda^k}{k!}$ | $\lambda$ | $\lambda$ |

## Continuous Distributions

| Distribution | Support | Parameters | PDF | Mean | Variance |
|---------------|----------|-------------|-----------|-----------|-----------|
| Uniform $(a,b)$ | $[a,b]$ | $a,b$ | $\frac{1}{b-a}$ | $\frac{a+b}{2}$ | $\frac{(b-a)^2}{12}$ |
| Exponential $(\lambda)$ | $[0,\infty)$ | $\lambda$ | $\lambda e^{-\lambda x}$ | $\frac{1}{\lambda}$ | $\frac{1}{\lambda^2}$ |
| Normal $(\mu,\sigma^2)$ | $(-\infty,\infty)$ | $\mu,\sigma$ | $\frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$ | $\mu$ | $\sigma^2$ |
| Gamma $(\alpha,\beta)$ | $[0,\infty)$ | $\alpha,\beta$ | $\frac{\beta^\alpha x^{\alpha-1} e^{-\beta x}}{\Gamma(\alpha)}$ | $\frac{\alpha}{\beta}$ | $\frac{\alpha}{\beta^2}$ |
| Beta $(\alpha,\beta)$ | $[0,1]$ | $\alpha,\beta$ | $\frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha,\beta)}$ | $\frac{\alpha}{\alpha+\beta}$ | $\frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}$ |
